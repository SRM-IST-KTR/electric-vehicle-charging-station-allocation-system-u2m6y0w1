```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb 
%matplotlib inline
plt.show()
```

```python
df = pd.read_csv('./OneDrive/Desktop/FP/electric_vehicle_charging_station_list.csv')
```

```python
df.drop(columns='Type',inplace=True)
```

```python
ax= plt.figure(figsize=(10,6))
sb.countplot(x='State',data=df, palette='hls')
plt.title('Number of charging stations in each State')
plt.xlabel('State')
plt.ylabel('Frequency')
```

```python
ax= plt.figure(figsize=(17,6))
sb.countplot(x='Address',data=df, palette='hls')
plt.title('Number of charging stations near a Facility')
plt.xlabel('Facility')
plt.ylabel('Frequency')
```

```python
ax= plt.figure(figsize=(15,6))
sb.countplot(x='Power',data=df, hue='State')
plt.title('Type of chargers in accordance to each State')
plt.xlabel('Type of charger')
plt.ylabel('frequency')
```

```python
ax= plt.figure(figsize=(15,6))
sb.countplot(x='State',data=df, hue='Address')
plt.title('Type of service in accordance to each State')
plt.xlabel('Type of charger')
plt.ylabel('frequency')
plt.legend(loc='center right', title='Address')
```
